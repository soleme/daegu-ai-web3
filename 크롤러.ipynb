### 크롤러의 완성

```python
def get_links(url):
    page = requests.get(url)
    soup = bs4.BeautifulSoup(page.text, 'html.parser')
    
    links = []
    for tag in soup.select('a'):
        if tag.get('href') is not None and tag['href'][:4] == 'http':
            links.append([tag['href'], tag.get_text()])

    return links
```

```python
def get_links(url):
    page = requests.get(url)
    soup = bs4.BeautifulSoup(page.text, 'html.parser')
    
    links = [[tag['href'], tag.get_text()] for tag in soup.select('a') if tag.get('href') is not None]
    links = [link for link in links if link[0][:4] == 'http']
    return links
```

```python
links = get_links('http://info.cern.ch')
print(links)
print(links[0][0]) # 첫번째 링크 - http://info.cern.ch/hypertext/WWW/TheProject.html

links += get_links(links[0][0])
print(links)
print(links[0][1]) # 첫번째 링크 - WhatIs.htm

# WhatIs.htm 는 웹 주소가 아니므로 여기서 에러
links += get_links(links[1][0])
print(links)
```

```python
links = get_links('http://info.cern.ch')

while len(links) > 0:
    link = links.pop(0)[0]
    new_links = get_links(link)
    print(link, new_links)
    links += new_links
    print(len(links))
```

```python
import requests
import bs4

def get_movies_from_daum(week=None):
    url = 'https://movie.daum.net/ranking/boxoffice/weekly'
    if week is not None:
        url = f'{url}?date={week}'
    page = requests.get(url)
    soup = bs4.BeautifulSoup(page.text, 'html.parser')
    movies = soup.select('.list_movieranking li')

    scrap = []
    for i, m in enumerate(movies):
        scrap.append({
            'week': week,
            'rank': i + 1,
            '제목': m.select('.link_txt')[0].get_text(),
            '개봉일': m.select('.info_txt .txt_num')[0].get_text(),
            '관객수': m.select('.info_txt')[1].get_text()[3:],
            '관람가능연령': m.select('.ico_movie')[0].get_text(),
            '설명': m.select('.link_story')[0].get_text().strip(),
            '이미지링크': m.select('.img_thumb')[0]['src'],
            '상세링크': m.select('.link_story')[0]['href']
        })

    return scrap
```

```python
from datetime import date

def get_week_list(week_count=1):
    weeks = []
    for i in range(week_count):
        week = date.today() - timedelta(weeks=(i + 1))
        weeks.append(week.strftime("%Y%m%d"))
    return weeks

print(get_week_list(3))
print(get_week_list(50))
```

```python
import requests
import bs4

def get_movies_from_daum(week=None):
    url = 'https://movie.daum.net/ranking/boxoffice/weekly'
    if week is not None:
        url = f'{url}?date={week}'
    page = requests.get(url)
    soup = bs4.BeautifulSoup(page.text, 'html.parser')
    movies = soup.select('.list_movieranking li')

    scrap = []
    for i, m in enumerate(movies):
        # 에러 대응 코드
        if len(m.select('.info_txt .txt_num')) == 0:
            continue
        if len(m.select('.img_thumb')) == 0:
            continue

        scrap.append({
            'week': week,
            'rank': i + 1,
            '제목': m.select('.link_txt')[0].get_text(),
            '개봉일': m.select('.info_txt .txt_num')[0].get_text(),
            '관객수': m.select('.info_txt')[1].get_text()[3:],
            '관람가능연령': m.select('.ico_movie')[0].get_text(),
            '설명': m.select('.link_story')[0].get_text().strip(),
            '이미지링크': m.select('.img_thumb')[0]['src'],
            '상세링크': m.select('.link_story')[0]['href']
        })

    return scrap
```

```python
movies = []
for w in get_week_list(50):
    movies += get_movies_from_daum(w)
```

```python
# 야구게임
# 3자리 숫자를 정해놓고 맞추는 게임. 자릿수가 맞으면 strike, 숫자만 맞추면 ball
import random
answer = random.sample("0123456789", k=3)

strike = 0
while strike != 3:
    inputs = input("세자리 숫자를 입력해주세요: ")
    strike = sum([a == b for a, b in zip(answer, inputs)])
    ball = sum([n in answer for n in inputs]) - strike
    print(f"{strike} strike, {ball} ball")

print("정답입니다.")
```
